```python
# Import packages
import numpy as np
import tomopy
import cv2
import matplotlib.pyplot as plt
from skimage.transform import rotate, AffineTransform
import tifffile
```

```python
# Creating ground truth tomography
data = tomopy.shepp3d(256)
ang = tomopy.angles(181)
proj = tomopy.project(data, ang)
```

```python
print(proj.shape)
recon = tomopy.recon(proj, ang, num_gridx = proj.shape[2], algorithm='gridrec', sinogram_order=False)
print(recon.shape)
reproj = tomopy.project(recon, ang, center = None, pad = False)
print(reproj.shape)
```

```python
# Sinogram for original
plt.imshow(proj[:, 128, :])
plt.title('Original')
plt.show()

# Sinogram for misaligned
plt.imshow(reproj[:, 128, :])
plt.title('Reprojected')
plt.show()
```
