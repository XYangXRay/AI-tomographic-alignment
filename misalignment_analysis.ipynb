# Misaligment Analysis
## Setup
Import packages for gathering and processing tomographic data.

```python
# Packages
import tomopy
import dxchange
import matplotlib.pyplot as plt
```

## Tomographic Data without Artificial Jitter
In order to understand the data and effect of jitter on the sinograms and tomographic reconstructions, start by using raw data to create a sinogram and reconstruction.

```python
# Generate an object with uniformly spaced scan angles and reconstruct with gridrec
obj = tomopy.shepp3d()
ang = tomopy.angles(180)
sim = tomopy.project(obj, ang)
rec = tomopy.recon(sim, ang, algorithm='gridrec')

# Plotting raw sinogram
plt.imshow(sim[:, 0, :])
plt.show()

# Show 64th slice of the reconstructed object.
plt.imshow(rec[64])
plt.show()
```

## Tomographic Data with Artificial Jitter
Now that the baseline has been established, artificial jitter will be added to the projection and the changes will be analyzed.

```python
# Generate an object with uniformly spaced scan angles and reconstruct with gridrec
obj = tomopy.shepp3d()
ang = tomopy.angles(180)

sim2 = tomopy.project(obj, ang)
tomopy.add_jitter(sim2, 1, 5) # add jitter
rec2 = tomopy.recon(sim2, ang, algorithm='gridrec')

# Plotting raw sinogram without jitter
plt.imshow(sim[:, 0, :])
plt.show()

# Plotting raw sinogram with jitter
plt.imshow(sim2[:, 0, :])
plt.show()

# Show 64th slice of the reconstructed object without jitter
plt.imshow(rec[64])
plt.show()

# Show 64th slice of the reconstructed object with jitter
plt.imshow(rec2[64])
plt.show()
```

## Tomographic Data Realigned with TomoPy Joint Alignment Algorithm
Now showing the sinogram and image after the joint alignment algorithm from TomoPy is used.

```python
tomopy.align_joint(sim2, ang, algorithm = 'gridrec')

# Plotting raw sinogram without jitter
plt.imshow(sim[:, 0, :])
plt.show()

# Plotting raw sinogram with jitter
plt.imshow(sim2[:, 0, :])
plt.show()

# Show 64th slice of the reconstructed object without jitter
plt.imshow(rec[64])
plt.show()

# Show 64th slice of the reconstructed object with jitter
plt.imshow(rec2[64])
plt.show()
```
