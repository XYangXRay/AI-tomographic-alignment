```python
import numpy as np
import tomopy
import matplotlib.pyplot as plt
import tifffile
```

```python
def nor_data(img):
#     mean_tmp = np.mean(img)
#     std_tmp = np.std(img)
#     img = (img - mean_tmp) / std_tmp
    img = (img - img.min())/(img.max()-img.min())
    return img
```

```python
prj = tifffile.imread('/data/prj_alignment/simulation/xz_noise01_aligned_miao.tiff')
prj = nor_data(prj)
ang = tomopy.angles(181)
center = tomopy.find_center(prj, ang)
extra_options ={'MinConstraint':0}
recon = np.zeros((200, 366, 366))
# tomopy.remove_nan(proj)
options = {'proj_type':'cuda', 'method':'EM_CUDA', 'num_iter':100, 'extra_options':extra_options}
for i in range(200):
    sino = prj[:,i,:].reshape((181,1,366))
    recon[i,:,:] = tomopy.recon(sino, ang, center=191, algorithm=tomopy.astra, options=options)
recon = tomopy.circ_mask(recon, axis=0, ratio=0.95)
tifffile.imsave('/data/prj_alignment/simulation/rec_xz_noise01_aligned_miao.tiff', recon)
```
